As we can see that thhis is an imbalanced dataset. So, we need to balance the dataset before the train test split

```python
X= df.iloc[:,:-1]
y = df.Class
```

```python
X.shape, y.shape
```

## Train test split

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.33, random_state=42)
```

```python
X_train.shape, y_train.shape
```

```python
X_test.shape, y_test.shape
```

## Balance Dataset

```python
from imblearn.combine import SMOTETomek
from imblearn.under_sampling import NearMiss
```

```python
smk = SMOTETomek(random_state=42)
```

```python
X_train, y_train = smk.fit_resample(X_train, y_train)
X_test, y_test = smk.fit_resample(X_test, y_test)
```

```python
X_train.shape, y_train.shape
```

```python
X_test.shape, y_test.shape
```

```python
cols = X_train.columns
```

```python
df1 = pd.concat([pd.DataFrame(X_train), pd.DataFrame(y_train)], axis=1)
df2 = pd.concat([pd.DataFrame(X_test), pd.DataFrame(y_test)], axis=1)
```

```python
balanced_df = pd.concat([df1,df2], axis=0)
```

```python
def max_depth_error(md):
    model = DecisionTreeClassifier(max_depth=md, random_state=42)
    model.fit(X_train, y_train)
    train_acc = 1 - model.score(X_train, y_train)
    val_acc = 1 - model.score(X_test, y_test)
    return {'Max Depth': md, 'Training Error': train_acc, 'Validation Error': val_acc}
```

```python
%%time

errors_df = pd.DataFrame([max_depth_error(md) for md in range(1, 21)])
```

```python
errors_df.sort_values(by = 'Validation Error')
```

```python
plt.figure()
plt.plot(errors_df['Max Depth'], errors_df['Training Error'])
plt.plot(errors_df['Max Depth'], errors_df['Validation Error'])
plt.title('Training vs. Validation Error')
plt.xticks(range(0,21, 2))
plt.xlabel('Max. Depth')
plt.ylabel('Prediction Error (1 - Accuracy)')
plt.legend(['Training', 'Validation'])
```

```python
def max_leaf_error(md):
    model = RandomForestClassifier(max_leaf_nodes=md ,n_estimators=400 , n_jobs = -1 ,random_state=42)
    model.fit(X_train, y_train);
    train_acc = 1 - model.score(X_train, y_train)
    val_acc = 1 - model.score(X_test, y_test)
    return {'Max leaf nodes': md, 'Training Error': train_acc, 'Validation Error': val_acc}
```
